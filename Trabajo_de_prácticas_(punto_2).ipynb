```python

print("DAMOS LA BIENVENIDA AL PROGRAMA...")

#Importamos las librerias
import RPI.GPIO as io
import time   
               
#Enlazamos el boton (button) al GPIO 10
button = 10               

#Enlazamos los Leds (A, B, C, D) al gpio (5, 7, 11, 13)
ledA = 5                    
ledB = 7                    
ledC = 11                    
ledD = 13                    


#Definimos
io.setmode(io.BCM)

#Evitamos señales de advertencias
io.setwarnings(False)        

#Defino los pines como salidas
io.setup(ledA,io.OUT)       
io.setup(ledB,io.OUT)        
io.setup(ledC,io.OUT)        
io.setup(ledD,io.OUT)        

#El button definido como entrada
io.setup(button,io.IN)     



while io.input(button):

#0000
    io.output(ledA,0)        
    io.output(ledB,0)
    io.output(ledC,0)
    io.output(ledD,0)    
    time.sleep(1)          

#0001
    io.output(ledA,0)        
    io.output(ledB,0)
    io.output(ledC,0)
    io.output(ledD,1)    
    time.sleep(1)   

#0010
    io.output(ledA,0)        
    io.output(ledB,0)
    io.output(ledC,1)
    io.output(ledD,0)    
    time.sleep(1)   

#0011
    io.output(led1,0)        
    io.output(led2,0)
    io.output(led3,1)
    io.output(led4,1)    
    time.sleep(1)   


#0100
    io.output(ledA,0)        
    io.output(ledB,1)
    io.output(ledC,0)
    io.output(ledD,0)    
    time.sleep(1)   


#0101
    io.output(ledA,0)        
    io.output(ledB,1)
    io.output(ledC,0)
    io.output(ledD,1)    
    time.sleep(1)   


#0110
    io.output(ledA,0)        
    io.output(ledB,1)
    io.output(ledC,1)
    io.output(ledD,0)    
    time.sleep(1)   


#0111
    io.output(ledA,0)        
    io.output(ledB,1)
    io.output(ledC,1)
    io.output(ledD,1)    
    time.sleep(1)   


#1000
    io.output(ledA,1)        
    io.output(ledB,0)
    io.output(ledC,0)
    io.output(ledD,0)    
    time.sleep(1)   


#1001
    io.output(ledA,1)        
    io.output(ledB,0)
    io.output(ledC,0)
    io.output(ledD,1)    
    time.sleep(1)
```
